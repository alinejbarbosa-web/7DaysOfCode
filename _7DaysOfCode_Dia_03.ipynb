```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arquivo_dia02 = "/content/drive/MyDrive/desafio_biblioteca/dia02_dados.csv"
dados = pd.read_csv(arquivo_dia02)
```

```python
dados['data_emprestimo'] = pd.to_datetime(dados['data_emprestimo'])
dados['data_devolucao'] = pd.to_datetime(dados['data_devolucao'])
```

```python
dados['ano'] = dados['data_emprestimo'].dt.year
dados['mes'] = dados['data_emprestimo'].dt.month
dados['hora'] = dados['data_emprestimo'].dt.hour
```

```python
def cdu_classe(localizacao):
    try:
        cod = int(localizacao)
        if 0 <= cod <= 99:
            return "Generalidades. Ciência e conhecimento"
        elif 100 <= cod <= 199:
            return "Filosofia e psicologia"
        elif 200 <= cod <= 299:
            return "Religião"
        elif 300 <= cod <= 399:
            return "Ciências sociais"
        elif 400 <= cod <= 499:
            return "Classe vaga"
        elif 500 <= cod <= 599:
            return "Matemática e ciências naturais"
        elif 600 <= cod <= 699:
            return "Ciências aplicadas"
        elif 700 <= cod <= 799:
            return "Belas artes"
        elif 800 <= cod <= 899:
            return "Linguagem. Língua. Linguística"
        elif 900 <= cod <= 999:
            return "Geografia. Biografia. História"
        else:
            return "Outros"
    except:
        return "Indefinido"

dados['classe_cdu'] = dados['localizacao'].apply(cdu_classe)
```

```python
import matplotlib.patches as mpatches
import matplotlib.lines as mlines # Import Line2D

emprestimos_por_biblioteca = dados.groupby('biblioteca')['id_exemplar'].count().sort_values(ascending=False)

plt.figure(figsize=(15, 6)) # Increased figure width
bars = emprestimos_por_biblioteca.plot(kind='bar', color='teal')
plt.title("Exemplares Emprestados por Biblioteca", fontsize=14)
plt.xlabel("Bibliotecas", fontsize=12) # Changed x-label to reflect numbers
plt.ylabel("Exemplares Emprestados", fontsize=12)
plt.xticks(ticks=range(len(emprestimos_por_biblioteca)), labels=range(1, len(emprestimos_por_biblioteca) + 1), rotation=0) # Set x-ticks to be numbers 1 to N
plt.grid(axis='y')

# Create custom legend handles and labels
legend_handles = []
legend_labels = []
for i, (library_name, count) in enumerate(emprestimos_por_biblioteca.items()):
    legend_labels.append(f"{i+1}: {library_name}")
    # Create a dummy handle (invisible line) for each legend entry
    legend_handles.append(mlines.Line2D([], [], color='teal', marker='s', markersize=0, linestyle='None'))


plt.legend(handles=legend_handles, labels=legend_labels, title="Bibliotecas", bbox_to_anchor=(1.05, 1), loc='upper left')


plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to make space for the legend
plt.show()
```

```python
print("➡️ Análise por biblioteca: identifica quais unidades têm maior demanda e precisam de mais recursos.")
print("➡️ Análise por Classe CDU (Classificação Decimal Universal): identifica quais áreas do conhecimento são mais procuradas pelos usuários.")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

emprestimos_por_cdu = dados.groupby('classe_cdu')['id_exemplar'].count().sort_values(ascending=False)

plt.figure(figsize=(12, 6)) # Increased figure height for better label display

# Generate a list of colors, one for each bar
colors = plt.cm.viridis(np.linspace(0, 1, len(emprestimos_por_cdu)))

bars = emprestimos_por_cdu.plot(kind='bar', color=colors)
plt.title("Exemplares Emprestados por Classe CDU", fontsize=14)
plt.xlabel("Classes CDU ", fontsize=12) # Changed x-label
plt.ylabel("Exemplares Emprestados", fontsize=12)
plt.xticks([]) # Remove x-axis ticks and labels
plt.grid(axis='y')

# Create custom legend handles and labels
legend_handles = []
legend_labels = []
for i, (cdu_class, count) in enumerate(emprestimos_por_cdu.items()):
    legend_labels.append(cdu_class)
    # Create a colored patch for each legend entry
    legend_handles.append(mpatches.Patch(color=colors[i], label=cdu_class))

plt.legend(handles=legend_handles, labels=legend_labels, title="Classes CDU", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to make space for the legend
plt.show()
```

```python
print("➡️ Análise por classe CDU: verifica os temas mais consultados pelos usuários.")
```
