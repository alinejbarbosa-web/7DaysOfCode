```python
import os

dfs = []
for file in os.listdir():
    if file.startswith("emprestimos-") and file.endswith(".csv"):
        df = pd.read_csv(file, encoding="utf-8")
        dfs.append(df)

emprestimos = pd.concat(dfs, ignore_index=True)
display(emprestimos.head())
```

```python
acervo = pd.read_parquet("dados_exemplares.parquet")
display(acervo.head())
```

```python
emprestimos['data_emprestimo'] = pd.to_datetime(emprestimos['data_emprestimo'], errors='coerce')
```

```python
ano_atual = pd.Timestamp.now().year
```

```python
emprestimos = emprestimos[emprestimos['data_emprestimo'].dt.year >= ano_atual - 10]
```

```python
dados = emprestimos.merge(acervo, on="codigo_barras", how="left")
```

```python
dados = dados.drop_duplicates()
```

```python
dados = dados.dropna(subset=["codigo_barras", "data_emprestimo"])
```

```python
display(dados.head())
print(f"Total de registros após limpeza: {len(dados)}")
```

```python
dados.to_csv("dia01_dados.csv", index=False, encoding="utf-8")
```

```python
dados.to_parquet("dia01_dados.parquet", index=False)
```

```python
acervo = pd.read_parquet("dados_exemplares.parquet")
display(acervo.head())
```
