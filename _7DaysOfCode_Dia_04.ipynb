```python
dados = pd.read_csv(arquivo_dia03)
```

```python
display(dados.head())
print(dados.info())
```

```python
def tabela_frequencia(df, coluna):
    freq = df[coluna].value_counts(dropna=False)
    perc = round(df[coluna].value_counts(normalize=True, dropna=False)*100, 2)
    tabela = pd.DataFrame({coluna+'_freq': freq, coluna+'_perc': perc})
    return tabela
```

```python
tabela_vinculo = tabela_frequencia(dados, 'tipo_vinculo_usuario')
display(tabela_vinculo)
```

```python
def cdu_classe(codigo):
    if pd.isna(codigo):
        return "Desconhecido"
    codigo = int(codigo)
    if 0 <= codigo <= 99:
        return "Generalidades"
    elif 100 <= codigo <= 199:
        return "Filosofia e psicologia"
    elif 200 <= codigo <= 299:
        return "Religião"
    elif 300 <= codigo <= 399:
        return "Ciências sociais"
    elif 400 <= codigo <= 499:
        return "Classe vaga"
    elif 500 <= codigo <= 599:
        return "Matemática e ciências naturais"
    elif 600 <= codigo <= 699:
        return "Ciências aplicadas"
    elif 700 <= codigo <= 799:
        return "Belas artes"
    elif 800 <= codigo <= 899:
        return "Linguagem"
    elif 900 <= codigo <= 999:
        return "Geografia, Biografia, História"
    else:
        return "Desconhecido"
```

```python
dados['classe_cdu'] = dados['localizacao'].apply(cdu_classe)
```

```python
colunas_categoricas = ['tipo_vinculo_usuario', 'colecao', 'biblioteca', 'classe_cdu']

tabelas = {}
for col in colunas_categoricas:
    tabelas[col] = tabela_frequencia(dados, col)
    print(f"✅ Tabela de frequência para {col}")
    display(tabelas[col])
```

```python
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches # Import mpatches for creating custom legend handles
import numpy as np # Import numpy for calculating ticks
import matplotlib.pyplot as plt # Import pyplot

sns.set(style="whitegrid")

# Plot for classe_cdu
col = 'classe_cdu'
plt.figure(figsize=(20, 10))

# Get the frequency of each class and sort it
counts = dados[col].value_counts().sort_index()

# Create a numerical mapping for the classes
mapping = {name: i for i, name in enumerate(counts.index)}
dados[f'{col}_numeric'] = dados[col].map(mapping)

# Create a vertical bar plot using the numeric representation
ax = sns.countplot(x=f'{col}_numeric', data=dados, order=mapping.values(), palette="viridis", hue=f'{col}_numeric', legend=False)
plt.title(f"Distribuição de empréstimos por {col}")
plt.xlabel(f"Código da {col}") # X-axis label now represents the numeric code
plt.ylabel("Quantidade de empréstimos") # Y-axis represents the count

# Set x-axis ticks to be the numeric values
ax.set_xticks(list(mapping.values()))
ax.set_xticklabels(list(mapping.values()))

# Set y-axis limits and ticks to show real numbers
max_count = counts.max()
ax.set_ylim(0, max_count * 1.1) # Set y-axis limit slightly above the max count
y_ticks = np.linspace(0, max_count * 1.1, 10) # Generate ticks
ax.set_yticks(y_ticks)
ax.yaxis.set_major_formatter(ticker.ScalarFormatter(useOffset=False, useMathText=False))


# Create custom legend handles based on the color palette and mapping
cmap = plt.cm.get_cmap("viridis", len(mapping)) # Get the colormap
legend_handles = [mpatches.Patch(color=cmap(i), label=f'{i}: {name}') for name, i in mapping.items()] # Create patches with colors from the colormap and labels from the mapping

plt.legend(handles=legend_handles, title=col, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent labels from being cut off and explicitly add legend if needed
plt.subplots_adjust(bottom=0.2, right=0.8) # Adjust the bottom and right margin for legend

plt.show()

# Drop the temporary numeric column if it exists
if f'{col}_numeric' in dados.columns:
    dados = dados.drop(columns=[f'{col}_numeric'])
```
